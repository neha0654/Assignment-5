```python
import numpy as np

# Define initial weights
W0 = 10
W1 = 0.2
W2 = -0.75
learning_rate = 0.05

# AND gate training data
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
Y = np.array([0, 0, 0, 1])

# Step activation function
def step_function(x):
    return 1 if x >= 0 else 0

# Track error values after each epoch
error_values = []

# Training the perceptron
epochs = 0
converged = False

while not converged:
    total_error = 0
    for i in range(len(X)):
        # Calculate the weighted sum
        weighted_sum = W0 + W1 * X[i][0] + W2 * X[i][1]
        # Apply step activation function
        predicted_output = step_function(weighted_sum)
        # Calculate error
        error = Y[i] - predicted_output
        total_error += error ** 2
        # Update weights
        W0 += learning_rate * error
        W1 += learning_rate * error * X[i][0]
        W2 += learning_rate * error * X[i][1]
    
    error_values.append(total_error)
    
    # Check for convergence
    if total_error == 0:
        converged = True
    epochs += 1

print(f"Converged in {epochs} epochs.")
print(f"Final weights: W0 = {W0}, W1 = {W1}, W2 = {W2}")
```

```python
import matplotlib.pyplot as plt
# Plot epochs vs. error values
plt.plot(range(1, epochs + 1), error_values)
plt.xlabel('Epochs')
plt.ylabel('Error')
plt.title('Convergence of Perceptron for AND Gate')
plt.show()
```
